```python
df = pd.read_csv("anspress_ekologiya_news.csv")
```

```python
df.head(30)
```

# Data Cleaning

```python
df.columns = df.columns.str.lower()
```

```python
df = df.drop([df.index[0], df.index[1], df.index[2]])
```

```python
df = df.drop("link", axis=1)
```

```python
columns = ["başlıq", "məqalə"]
df[columns] = df[columns].apply(lambda x: x.str.lower())
```

```python
df[columns] = df[columns].apply(lambda x: x.str.replace(r'\d', "", regex=True))
```

```python
df[columns] = df[columns].apply(lambda x: x.str.replace(r'[^\w\s]', "", regex = True))
```

```python
df[columns] = df[columns].apply(lambda x: x.str.replace(r'\n+', ' ', regex=True))
```

```python
df[columns] = df[columns].apply(lambda x: x.str.replace(r'\s+', ' ', regex=True).str.strip())
```

```python
remove_words = ['ans', 'press', 'xəbərlər', 'ekologiya', 'feb', 'jan', 'dec', 'sep', 'oct', 'nov', 'yeniləndi', 'və', 'bu', 'ki', 'ilə', 'üçün', 'müddəti', 
               'oxuma', 'kopyala', 'barədə', 'edilib', 'də', 'bildirib', 'aşkar', 'bağlı', 'görə', 'bir', 'üzrə', 'dəqiqədən', 'az', 'o', 'dəqiqə', 'onun', 'olunan', 'olan', 'qeyd', 'edib', 'hazırda',
               'isə']
df[columns] = df[columns].apply(lambda x: x.str.replace(r'\b(' + '|'.join(remove_words) + r')\b', '', regex=True).str.replace(r'\s+', ' ', regex=True).str.strip())
```

```python
tf = df['məqalə'].dropna().str.split().explode().value_counts().reset_index()
```

```python
tf.columns = ['word', 'count']
```

```python
tf = tf[tf["count"] >= 2]
```

```python
tf.sort_values("count", ascending = False)
```
